#FotMob(individual stats)

```python
from selenium import webdriver
from selenium.webdriver.support.select import Select
from selenium.webdriver.chrome.options import Options
import pandas as pd

DRIVER_PATH = '/usr/bin/chromedriver'

chrome_options = Options()
chrome_options.binary_location = DRIVER_PATH
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("--window-size=2560,1440")

driver = webdriver.Chrome(options=chrome_options)
driver.get('https://www.fotmob.com/matches/-vs-/37q8pa#4314987')
```

```python
from selenium.webdriver.common.by import By
```

```python
driver.find_elements(By.CLASS_NAME, "css-jspn1-FilterButtonsContainer")[1].find_elements(By.TAG_NAME, "button")[2].click()
driver.find_elements(By.CLASS_NAME, "css-17lsavb-SeeMoreContainer")[0].find_element(By.TAG_NAME, "button").click()

table = driver.find_elements(By.CLASS_NAME, "css-zh2ah-StyledTable")[0].text
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# Initialize an empty DataFrame with columns
columns = ['Player', 'Touches', 'Accurate passes', 'Accurate passes(%)', 'Assists', 'xA', 'Chances created', 'Passes into final third', 'Accurate crosses', 'Accurate crosses(%)', 'Accurate long balls', 'Accurate long balls(%)']
df = pd.DataFrame(columns=columns)

index_toby = table.find("Toby Alderweireld")
splitted = table[index_toby:].split("\n")

for i in range(0, 54, 2):
  new_row = {}
  new_row['Player'] = splitted[i]
  splitted_stats = splitted[i+1].split(" ")

  new_row['Touches'] = splitted_stats[0]
  new_row['Accurate passes'] = splitted_stats[1]
  new_row['Accurate passes(%)'] = splitted_stats[2]
  new_row['Assists'] = splitted_stats[3]
  new_row['xA'] = splitted_stats[4]
  new_row['Chances created'] = splitted_stats[5]
  new_row['Passes into final third'] = splitted_stats[6]
  if splitted_stats[7] == '0':
      new_row['Accurate crosses'] = splitted_stats[7]
      new_row['Accurate crosses(%)'] = '-'
      if splitted_stats[8] == '0':
        new_row['Accurate long balls'] = splitted_stats[8]
        new_row['Accurate long balls(%)'] = '-'
      else:
        new_row['Accurate long balls'] = splitted_stats[8]
        new_row['Accurate long balls(%)'] = splitted_stats[9]
  else:
      new_row['Accurate crosses'] = splitted_stats[7]
      new_row['Accurate crosses(%)'] = splitted_stats[8]
      if splitted_stats[9] == '0':
        new_row['Accurate long balls'] = splitted_stats[9]
        new_row['Accurate long balls(%)'] = '-'
      else:
        new_row['Accurate long balls'] = splitted_stats[9]
        new_row['Accurate long balls(%)'] = splitted_stats[10]


  df = df._append(new_row, ignore_index=True)
```

#UEFA (team stats)

```python
from selenium import webdriver
from selenium.webdriver.support.select import Select
from selenium.webdriver.chrome.options import Options
import pandas as pd

DRIVER_PATH = '/usr/bin/chromedriver'

chrome_options = Options()
chrome_options.binary_location = DRIVER_PATH
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("--window-size=2560,1440")

driver_uefa = webdriver.Chrome(options=chrome_options)



driver_uefa.get('https://www.uefa.com/uefachampionsleague/match/2039107--antwerp-vs-shakhtar-donetsk/statistics/')
```

```python
sraped_uefa_stats = driver_uefa.find_elements(By.XPATH, '//*[@id="accordion-content-4"]//div')[0].text.split("\n")
uefa_passing = dict()
uefa_passing['Team'] = ['Antwerp', 'Shakhtar']
for i in range(0,54,3):
  uefa_passing[sraped_uefa_stats[i+1]] = [sraped_uefa_stats[i], sraped_uefa_stats[i+2]]

df_uefa = pd.DataFrame(uefa_passing)
df_uefa
```
